```python
# Load the CSV file into a DataFrame
csv_file = "log_24_2024-8-21-17-27-44.csv"  # Replace with your file path
df = pd.read_csv(csv_file)

# Display the first few rows
print(df.head())
```

```python
# General info about the dataset
df.info()

# Summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Get the unique mavpackettypes
packet_types = df['mavpackettype'].unique()

print("Unique mavpackettypes in the DataFrame:")
print(packet_types)
```

```python
# Create a dictionary to hold DataFrames for each mavpackettype
message_dfs = {}

# Group the DataFrame by 'mavpackettype'
grouped = df.groupby('mavpackettype')

# Iterate over each group
for packet_type, group_df in grouped:
    # Reset the index of the group DataFrame
    group_df = group_df.reset_index(drop=True)
    
    # Add the group DataFrame to the dictionary
    message_dfs[packet_type] = group_df
    
    # Print information about the group
    print(f"DataFrame for '{packet_type}': {group_df.shape[0]} rows, {group_df.shape[1]} columns")
```

```python
# Create an empty DataFrame to hold the combined result
combined_df = pd.DataFrame()

# Iterate over each message type DataFrame
for packet_type, message_df in message_dfs.items():
    # Identify columns where all values are null
    all_null_columns = message_df.columns[message_df.isnull().all()].tolist()
    
    # Drop all-null columns
    message_df_cleaned = message_df.drop(columns=all_null_columns)
    
    # Add a column to identify the source packet type (optional)
    message_df_cleaned['mavpackettype'] = packet_type

    # Merge the cleaned DataFrame into the combined DataFrame
    combined_df = pd.concat([combined_df, message_df_cleaned], axis=0, ignore_index=True)
    
    # Print information about the cleaned DataFrame
    print(f"\n'{packet_type}' cleaned:")
    print(f"Dropped {len(all_null_columns)} all-null columns.")
    print(f"Remaining columns ({message_df_cleaned.shape[1]}): {message_df_cleaned.columns.tolist()}")

# Display the combined DataFrame structure
print("\nCombined DataFrame:")
print(combined_df.info())
print(combined_df.head())
```

```python
# Display general information about the DataFrame
print("General Information:")
print(combined_df.info())

# Display basic statistics for numeric columns
print("\nBasic Statistics:")
print(combined_df.describe())

# Count null values for each column
print("\nNull Value Counts:")
null_counts = combined_df.isnull().sum()
print(null_counts)

# Calculate the percentage of missing values for each column
print("\nPercentage of Missing Values:")
missing_percentage = (combined_df.isnull().mean() * 100).sort_values(ascending=False)
print(missing_percentage)

# Display columns with more than 50% missing values
print("\nColumns with >50% Missing Values:")
high_null_columns = missing_percentage[missing_percentage > 50]
print(high_null_columns)

# Display columns with 0% missing values
print("\nColumns with No Missing Values:")
no_null_columns = missing_percentage[missing_percentage == 0]
print(no_null_columns)

# Identify duplicate rows (optional)
duplicate_count = combined_df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")
```

```python
# Drop columns with only one unique value
constant_columns = [col for col in combined_df.columns if combined_df[col].nunique() == 1]
combined_df = combined_df.drop(columns=constant_columns)

print(f"Columns dropped due to constant values: {constant_columns}")
```
